Import libraries and load the data table

```python
import pandas as pd
import time
import numpy as np

df = pd.read_csv('BMD_GO_Analysis_COMPLETE_Combined Analyses_filtered.txt', 
                 sep="\t", header=0,  encoding='latin-1', low_memory=False)
df.head(n=5)
```

Filter for FDR values and number of genes

```python
df1 = df.loc[df['Genes That Passed All Filters'] > 5]
df1 = df.loc[df['Fisher\'s Exact Two Tail'] < 0.05]
#df1 = df.loc[df['GO Level'] == 2]
df1.head()
```

```python
chems = df['Analysis'].unique()
chems
```

Below, the lowest 20 BMD pathways are extracted

```python
for i in chems:
    #print(i)
    df2 = df1.loc[(df1['Analysis'] == i)] # | (df1['GO Level'] == 'Ziram') ]

    df2.sort_values(by=['BMD Mean'],  ascending=True)

#df3 = df2.loc[(df2['BMD Mean'])].notna()
    #df2 = df2['Fisher\'s Exact Two Tail'].sort_values(axis=1, ascending=True)
#file = "test.txt"
#df2.to_csv(file, sep="\t")
    df3 = df2.dropna(subset = ['BMD Mean'])
    df4 = df3.iloc[:20]
    file= "Lowest20pathways_meanBMD_perPathway_" + str(i) + ".txt"
    df4.to_csv(file, sep="\t")
```

```python
import os
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
relevant_path = "./"
included_extensions = ['txt']
file_names = [fn for fn in os.listdir(relevant_path)
              if any(fn.startswith("Lowest10pathways") for ext in included_extensions)]


genes_bmd = pd.read_csv('Combined Analyses_filtered_BenchmarkDosing_Genes_COMPLETE.txt', 
                 sep="\t", header=0,  encoding='latin-1', low_memory=False)

# Create an empty dataframe to store the extracted columns
merged_df = pd.DataFrame()
bmd_column = pd.DataFrame()
# Loop through all files in the directory with the given filename pattern

genel=list()
chemgenedic=dict()
for i in file_names:
        import re

        text = str(i)

        m = re.search('Lowest10pathways_meanBMD_perPathway_(.+?).txt', text) ### Chane the # of pathways here
        if m:
            found = m.group(1)
            #print(found)
            
            
        data = pd.read_csv(i, header=None, sep="\t")
        #data.columns = data.index[1]
        data.columns = data.iloc[0]
        data = data.iloc[1:].reset_index(drop=True)
        
        to_extract = data['Gene Symbols']
        
       
        #genel1 = pd.DataFrame()
        
        chemgenedic[str(found)] = ""
        
     
        for row in to_extract: 
                gene_list=row.split(";")
                for gene in gene_list:
                        if gene not in genel: genel.append(gene)
                        if gene not in chemgenedic[str(found)]: 
                            chemgenedic[str(found)]=chemgenedic[str(found)] + "," + gene
                    
#print(genedic)
```

```python
genedic=dict()
chemdic=dict()
setdic=dict()
it=0
with open("Combined Analyses_filtered_BenchmarkDosing_Genes_COMPLETE.txt") as f:
    for line in f:
        it=it+1
        if it>1:
            nline=line.strip()
            nlinel=nline.split()
            ch = nlinel[0]
            gene=nlinel[3]
            setc=ch+gene
            setcval=nlinel[67]
            if ch not in chemdic: chemdic[ch]=gene
            elif ch in chemdic and gene not in chemdic[ch] :  chemdic[ch] =  chemdic[ch] + "," + gene
            if gene not in genedic: genedic[gene]=ch
            elif gene in genedic and ch not in genedic[gene] : genedic[gene] = genedic[gene] + "," + ch
            setdic[setc]=setcval        
fileout=open("Lowest10pathways_BestBMD_PERGENE.txt", "w")
fileout.write("genes" + "\t")
for it in chemdic: fileout.write(it + "\t")
fileout.write("\n")
for it1 in genel:
    fileout.write(it1)
    for it2 in chemdic:
        if it2 in chemgenedic and it1 in chemgenedic[it2] and it1 in genedic and it2 in genedic[it1]:
            valkey=it2+it1
            if valkey in setdic : fileout.write("\t" + str(setdic[valkey]))
        else : 
            fileout.write("\t" + "NA")
    fileout.write("\n")


```
